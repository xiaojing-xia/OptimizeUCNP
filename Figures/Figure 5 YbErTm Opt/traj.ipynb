```python
import argparse
from datetime import datetime
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time
import torch

plt.rcParams["font.family"] = "arial"

DATA_DEST1 = "../saved_data/simulation_log_YbErTm_VIStoUV_logEmission_beta=5_10inits_0.csv"
DATA_DEST2 = "../saved_data/simulation_log_YbErTm_VIStoUV_logEmission_beta=5_10inits_1.csv"
DATA_DEST3 = "../saved_data/simulation_log_YbErTm_VIStoUV_logEmission_beta=5_10inits_2.csv"
DATA_DEST4 = "../saved_data/simulation_log_YbErTm_VIStoUV_logEmission_beta=5_10inits_3.csv"


df1 = pd.read_csv(DATA_DEST1)
df2 = pd.read_csv(DATA_DEST2)
df3 = pd.read_csv(DATA_DEST3)
df4 = pd.read_csv(DATA_DEST4)



def feature_traj(df,feature_idx):
    evalued = len(df)
    y_ = []
    x_ = []
    for i in range(evalued):
        y_.append(df.iloc[i,feature_idx])
        x_.append(i)
    return x_, y_

def get_Max(df, feature_idx):
    evalued = len(df)
    max_feature = []
    for i in range(evalued):
        max_feature.append(df.iloc[df.iloc[0:i+1].UV.idxmax(),feature_idx])
    x = range(len(max_feature)) 
    return x, max_feature
```

```python

def plot_traj(df, df_name, ft_idx, lgd_name, sav_name, color):
    FontSize = 18
    ticksize = 14
    x1,yb1 = feature_traj(df, ft_idx)

    fig, ax1 = plt.subplots(figsize=(8,6))
    plt.plot(x1,[i *100 for i in yb1], c= color, label = lgd_name)
    x11,y11 = get_Max(df,ft_idx)
    plt.plot(x11,[i *100 for i in y11], c= 'grey', linestyle = 'dashed', label = 'Best '+lgd_name)

    plt.legend(fontsize = FontSize, frameon=False)
    plt.axvspan(0, 10, color='grey', alpha=0.19)
    plt.xlabel('# Evaluated NPs', fontsize = FontSize)
    plt.ylabel('Doping Concentration (%)', fontsize = FontSize)
    #plt.xlim([0,48])
    plt.ylim([-5,105])
    plt.locator_params(axis='y', nbins=4)
    plt.xticks(fontsize=ticksize)
    plt.yticks(fontsize=ticksize)
    plt.legend(loc='upper left', fontsize = FontSize,frameon=False)
    #plt.savefig('YbErTm_opt_progress_obj=VIS.eps') 
    plt.savefig('SI_opt_progress_'+sav_name+'_'+df_name+'.pdf') 
    
def plot_traj_radius(df, df_name, ft_idx):
    FontSize = 18
    ticksize = 14
    x1,yb1 = feature_traj(df, ft_idx)

    fig, ax1 = plt.subplots(figsize=(8,6))
    plt.plot(x1,[i /10 for i in yb1], c= 'black', label = 'Core radius')
    x11,y11 = get_Max(df,ft_idx)
    plt.plot(x11,[i/10 for i in y11], c= 'grey', label = 'Best radius')

    plt.legend(fontsize = FontSize, frameon=False)
    plt.axvspan(0, 10, color='grey', alpha=0.19)
    plt.xlabel('# Evaluated NPs', fontsize = FontSize)
    plt.ylabel('Core Radius (nm)', fontsize = FontSize)
    #plt.xlim([0,48])
    plt.ylim([0,5])
    plt.locator_params(axis='y', nbins=4)
    plt.xticks(fontsize=ticksize)
    plt.yticks(fontsize=ticksize)
    plt.legend(loc='upper left', fontsize = FontSize,frameon=False)
    #plt.savefig('YbErTm_opt_progress_obj=VIS.eps') 
    plt.savefig('SI_opt_progress_'+sav_name+'_'+df_name+'.pdf') 
```

```python
df = df1
df_name = 'df1'

c_yb = '#D5BB67'
c_er = '#E3A39A'
c_tm = '#78ACA8'

ft_idx = 0
lgd_name = '$Yb_1$'
sav_name = 'Yb1'
c = c_yb
plot_traj(df, df_name, ft_idx, lgd_name, sav_name, c)

ft_idx = 1
lgd_name = '$Er_1$'
sav_name = 'Er1'
c = c_er
plot_traj(df, df_name, ft_idx, lgd_name, sav_name, c)

ft_idx = 2
lgd_name = '$Tm_1$'
sav_name = 'Tm1'
c = c_tm
plot_traj(df, df_name, ft_idx, lgd_name, sav_name, c)

ft_idx = 3
lgd_name = '$Yb_2$'
sav_name = 'Yb2'
c = c_yb
plot_traj(df, df_name, ft_idx, lgd_name, sav_name, c)

ft_idx = 4
lgd_name = '$Er_2$'
sav_name = 'Er2'
c = c_er
plot_traj(df, df_name, ft_idx, lgd_name, sav_name, c)

ft_idx = 5
lgd_name = '$Tm_2$'
sav_name = 'Tm2'
c = c_tm
plot_traj(df, df_name, ft_idx, lgd_name, sav_name, c)

ft_idx = 6
sav_name = 'radius'
plot_traj_radius(df, df_name, ft_idx)
```
